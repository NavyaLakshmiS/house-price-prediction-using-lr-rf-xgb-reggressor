```python
X = salary.iloc[:, :-1].values    # Features => Years of experience => Independent Variable
y = salary.iloc[:, -1].values     # Target => Salary => Dependent Variable
```

```python
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
predictions
```

```python
y_test
```

```python
sns.distplot(predictions-y_test)
```
